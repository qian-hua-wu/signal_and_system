```python
import numpy as np
import matplotlib.pyplot as pl
import pandas as pd
import scipy.signal as sg
```

```python
a= np.array([1, 1, 1, 1])
v= np.array([1, 1])

M= a.size
N= v.size

aa= np.concatenate((a,np.zeros(N-1)))
vv= np.concatenate((v,np.zeros(M-1)))

M= aa.size
N= vv.size

m= np.arange(M).reshape(1,-1)
n= np.arange(N).reshape(-1,1)

m, aa[m], n-m, vv[n-m]

aa[m]*vv[n-m]
```

```python
x = a
w = v
y = aa[m]*vv[n-m]

xk = np.fft.fft(x)
wk = np.fft.fft(w)
yk = np.fft.fft(y)

M= xk.size
N= wk.size

aa= np.concatenate((xk,np.zeros(N-1)))
vv= np.concatenate((wk,np.zeros(M-1)))

M= aa.size
N= vv.size

m= np.arange(M).reshape(1,-1)
n= np.arange(N).reshape(-1,1)

m, aa[m], n-m, vv[n-m]

print(aa[m]*vv[n-m])
print()
print(xk)
print(wk)
print(yk)
```
